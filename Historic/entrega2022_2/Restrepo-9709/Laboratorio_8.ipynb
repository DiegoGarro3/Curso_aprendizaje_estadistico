```python
# Librer√≠as 
import pandas as pd
import numpy as np
import h5py
import matplotlib.pyplot as plt
from tensorflow import keras
```

#### Se cargan los datos

```python
data_train= "Curso_aprendizaje_estadistico/Assesment/dataset/train_catvnoncat.h5"
train_dataset = h5py.File(data_train, "r")
 
data_test= "Curso_aprendizaje_estadistico/Assesment/dataset/test_catvnoncat.h5"
test_dataset = h5py.File(data_test, "r")
```

```python
#test_dataset.keys()
```

```python
# Read the data
xtrain_classes, xtrain, train_label =\
train_dataset["list_classes"],train_dataset["train_set_x"],train_dataset["train_set_y"]

test_classes, xtest,test_label =\
test_dataset["list_classes"],test_dataset["test_set_x"],test_dataset["test_set_y"]
```

```python
plt.imshow(xtrain[200])
print(train_label[200])
```

```python
# Se vectorizan los datos
xtrainv = np.reshape(xtrain,(209, 64*64*3))/255
xtestv = np.reshape(xtest,(50, 64*64*3))/255

train_labelv = np.reshape(train_label,209)
test_labelv = np.reshape(test_label,50)
```

```python
#xtrain_.shape
#xtest_.shape
```

#### Modelo de red neuronal (NN)

```python
#  Topolog√≠a de la red: [12288, 100, 1] con la funci√≥n de activaci√≥n relu y en la
#  √∫ltima capa la softmax

model = keras.models.Sequential()
model.add(keras.layers.Dense(100, input_dim=12288, activation='sigmoid'))
#model.add(keras.layers.Dense(100, activation = 'relu'))
#model.add(keras.layers.Dense(5, activation = 'relu'))
model.add(keras.layers.Dense(1, activation = 'softmax'))
```

```python
model.summary()
```

```python
# Se especifica la funci√≥n de coste, el optimizador y la m√©trica
opt = keras.optimizers.SGD(lr=0.01)
model.compile(loss='binary_crossentropy',optimizer=opt, metrics=['accuracy'])
```

```python
#, validation_data=(xtestv, test_labelv)
```

```python
# Se entrena la red neuronal
keras.backend.clear_session()
history = model.fit(xtrainv, train_labelv , validation_data=(xtestv, test_labelv),  epochs= 200, verbose=2)
```

```python
# Evaluaci√≥n del modelo
_, train_acc = model.evaluate(xtrainv, train_labelv, verbose=0)
_, test_acc = model.evaluate(xtestv, test_labelv, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))

# Gr√°ficas:
plt.figure(figsize=(14,7))

# Gr√°fica de p√©rdidas durante el entrenamiento
plt.subplot(121)
plt.title('Loss')
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()

# Gr√°fica de las precisi√≥n durante el entrenamiento
plt.subplot(122)
plt.title('Accuracy')
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='test')
plt.legend()
plt.show()
```

```python
# Creaci√≥n del modelo
def build_model2(activation = 'softplus', loss='binary_crossentropy'):
    model = keras.models.Sequential()
    model.add(keras.layers.Dense(10,input_dim=12288, activation='relu'))
    model.add(keras.layers.Dense(1, activation = activation))
    opt = keras.optimizers.SGD(lr=0.005)
    model.compile(loss=loss, optimizer=opt, metrics=['accuracy'])
    return model
```

```python
# binary cross entropy:

keras.backend.clear_session()


model = build_model2()
history = model.fit(xtrainv, train_labelv, validation_data=(xtestv, test_labelv), epochs=300, verbose=0)

# Evaluaci√≥n del modelo
_, train_acc = model.evaluate(xtrainv, train_labelv, verbose=0)
_, test_acc = model.evaluate(xtestv, test_labelv, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))

# Gr√°ficas:
plt.figure(figsize=(14,7))

# Gr√°fica de p√©rdidas durante el entrenamiento
plt.subplot(121)
plt.title('Loss')
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()

# Gr√°fica de las precisi√≥n durante el entrenamiento
plt.subplot(122)
plt.title('Accuracy')
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='test')
plt.legend()
plt.show()
```
