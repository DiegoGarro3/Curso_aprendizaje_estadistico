```python
%pylab inline
from scipy.stats import norm
import pandas as pd
from scipy import optimize
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import learning_curve
```

## Intuición Estadística.

Empezamos por construir dos distribuciones de probabilidad normales como medias $\mu_1=5.0$, $\mu_2=9.5$ y esviaciones estándar $\sigma_1=2.5$, $\sigma_2=1.5$.

```python
μ1, μ2, σ1, σ2 = 5.0, 9.5, 2.5, 1.5

dist1 = norm(loc = μ1, scale = σ1)
dist2 = norm(loc = μ2, scale = σ2)
```

A continuación se muestra un gráfico de ambas distribuciones.

```python
x = np.linspace(0,15,100)
plt.plot(x,dist1.pdf(x),label="Dist 1")
plt.plot(x,dist2.pdf(x),label="Dist 2")
plt.legend()
plt.grid()
```

Para separar los datos, podemos definir una **frontera analítica** como el punto de intersección de las dos funciones, de modo que 

$$\frac{1}{\sigma_1\sqrt{2\pi}}\exp\left[-\frac{1}{2}\left(\frac{x-\mu_1}{\sigma_1}\right)^2\right]=\frac{1}{\sigma_2\sqrt{2\pi}}\exp\left[-\frac{1}{2}\left(\frac{x-\mu_2}{\sigma_2}\right)^2\right]$$

$$\frac{\sigma_2}{\sigma_1}\exp\left[-\frac{1}{2}\left(\frac{x-\mu_1}{\sigma_1}\right)^2\right]=\exp\left[-\frac{1}{2}\left(\frac{x-\mu_2}{\sigma_2}\right)^2\right]$$

$$\frac{1}{2}\log\left[\left(\frac{\sigma_2}{\sigma_1}\right)^2\right]-\frac{1}{2}\left(\frac{x-\mu_1}{\sigma_1}\right)^2=-\frac{1}{2}\left(\frac{x-\mu_2}{\sigma_2}\right)^2$$

$$\log\left[\left(\frac{\sigma_2}{\sigma_1}\right)^2\right]+\left(\frac{x-\mu_2}{\sigma_2}\right)^2-\left(\frac{x-\mu_1}{\sigma_1}\right)^2=0$$

$$\log\left[\left(\frac{\sigma_2}{\sigma_1}\right)^2\right]+\frac{x^2-2\mu_2 x+\mu_2^2}{\sigma_2^2}-\frac{x^2-2\mu_1 x+\mu_1^2}{\sigma_1^2}=0$$

$$\log\left[\left(\frac{\sigma_2}{\sigma_1}\right)^2\right]+\frac{\mu_2^2}{\sigma_2^2}-\frac{\mu_1^2}{\sigma_1^2}+\left( \frac{2\mu_1}{\sigma_1^2}-\frac{2\mu_2}{\sigma_2^2}\right)x+\left(\frac{1}{\sigma_2^2}-\frac{1}{\sigma_1^2}\right)x^2=0$$

$$C + Bx + Ax^2 = 0$$

Con soluciones

$$x = \frac{-B \pm\sqrt{B^2-4AC}}{2A}$$

La implementación de esta expresión se encuentra en la siguiente función, además de que se añade el gráfico de la misma a la de las distribuciones.

```python
def Boundary(μ1, μ2, σ1, σ2):
  A = 1/σ2**2-1/σ1**2
  B = 2*μ1/σ1**2-2*μ2/σ2**2
  C = 2*np.log(σ2/σ1)+(μ2/σ2)**2-(μ1/σ1)**2
  return np.roots([A,B,C])
```

```python
xBound = Boundary(μ1, μ2, σ1, σ2)
x = np.linspace(0,15,100)
plt.plot(x,dist1.pdf(x),label="Dist 1")
plt.plot(x,dist2.pdf(x),label="Dist 2")
plt.vlines(xBound[0],0,0.25,linestyles='dotted')
plt.vlines(xBound[1],0,0.25,linestyles='dotted')
plt.legend()
```

## Modelo de Machine Learning.

En primer lugar, construimos un **conjunto de entrenamiento** al generar $100$ muestras aleatorias basados en las distribuciones ya creadas.

```python
sample1 = dist1.rvs(size = 100)
sample2 = dist2.rvs(size = 100)
```

```python
plt.hist(sample1,density=True)
plt.plot(x,dist1.pdf(x),label="Dist 1")
plt.hist(sample2,density=True)
plt.plot(x,dist2.pdf(x),label="Dist 2")
plt.legend()
```

Vamos a **etiquetar** a los números obtenidos de la distribución $1$ con $0$, y a los de la distribución $2$ con la etiqueta $1$. Y construimos un dataframe.

```python
X = np.zeros(200)
X[:100] = sample1
X[100:] = sample2

Y = np.zeros(200)
Y[100:] = 1
```

```python
dic = {'X1':X, 'Y':Y} 
data = pd.DataFrame(dic)
```

```python
data.head()
```

### Construyendo conjunto de prueba.

De la misma manera, vamos a construir un conjunto más pequeño, de $50$ mustras por distribución, para hacer **pruebas** del modelo.

```python
test_sample1 = dist1.rvs(size = 50)
test_sample2 = dist2.rvs(size = 50)
```

```python
X_test = np.zeros(100)
X_test[:50] = test_sample1
X_test[50:] = test_sample2

Y_test = np.zeros(100)
Y_test[50:] = 1

dic_test = {'X1t':X_test, 'Yt':Y_test} 
data_test = pd.DataFrame(dic_test)
```

```python
data_test.head()
```

## Contrucción del modelo.

Con `clf` se define el modelo, en este caso *LinearRegression*, con `clf.fit` se **entrena** el modelo con los datos definidos como conjunto de entrenamiento.

```python
X_train = data['X1'].to_numpy().reshape(-1,1)
Y_train = data['Y']

clf = LinearRegression()
clf.fit(X_train, Y_train)
```

Una primera opción para evaluar la efectividad del modelo es usar `score`.

```python
score_train = clf.score(X_train, Y_train)
score_train
```

Vemos que para el conjunto de entramiento dado la clasificación no tiene unos buenos resultados, pues el *score* es $0.65$. Si se evalúa este *score* sobre el conjunto de prueba, podemos obtener una cantidad que mida la capacidad del modelo para generalizar su aprendizaje.

```python
X_test = data_test['X1t'].to_numpy().reshape(-1,1)
Y_test = data_test['Yt']

score_test = clf.score(X_test, Y_test)
score_test
```

* Por el *score* medio que se obtuvo para *training_set* indica que el modelo no funciona bien para la clasificación de los números de acuerdo a las dos dstribuciones, poblemente debido a que el área en la que se superponen es grande y algunos datos de entrenamiento pertenecen a dicha región.

* El resultado tan bajo en el *test_set* puede decirnos que para datos que el modelo no ha "visto", la clasificación es errada, en otras palabras, el modelo no generaliza el entrenamiento dado.

```python
tree_clf = DecisionTreeClassifier(max_depth=3)
svm_clf = SVC(kernel="linear", C=0.025)
```

En comparación con el modelo de *LinearREgression*, los nuevos modelos **son mejores** en la tarea de clasificar el *training_set* y se manejan mucho mejor al generalizar para el *test_set*, habiendo ligeras diferencias entre ambos. 

### Optimización de hiperparámetros.

LOs modelos vistos requieren del ajuste de algunos parámetros, estos son llamados **hiperpaŕametros** del modelo, y existen formas de obtener los valores más adecuados para ellos.

```python
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5)
CV_rfc.fit(X_train, Y_train)
```

```python
params = CV_rfc.best_params_
```

Los parámetros que se obtienen usando la librería de optimización son los siguientes.

```python
params
```

```python
rfc_opt = RandomForestClassifier(**params)
rfc_opt.fit(X_train, Y_train)
print('Scores (Train / Test) para parámetros optimizados.')
print(*[rfc_opt.score(X_train, Y_train),rfc_opt.score(X_test, Y_test)], sep='\n')
```

Aunque para el *trainnig_set* el resultado es prácticamente $1$, se pierde algo de generalización para el *test_set*.

### Curvas de Aprendizaje.

Otra manera de evaluer al modelo es a través de las curvas de aprendizaje.

```python
def plot_learning_curve(
    estimator,
    title,
    X,
    y,
    axes=None,
    ylim=None,
    cv=None,
    n_jobs=None,
    train_sizes=np.linspace(0.1, 1.0, 5),
):
    """
    Generate 3 plots: the test and training learning curve, the training
    samples vs fit times curve, the fit times vs score curve.

    Parameters
    ----------
    estimator : estimator instance
        An estimator instance implementing `fit` and `predict` methods which
        will be cloned for each validation.

    title : str
        Title for the chart.

    X : array-like of shape (n_samples, n_features)
        Training vector, where ``n_samples`` is the number of samples and
        ``n_features`` is the number of features.

    y : array-like of shape (n_samples) or (n_samples, n_features)
        Target relative to ``X`` for classification or regression;
        None for unsupervised learning.

    axes : array-like of shape (3,), default=None
        Axes to use for plotting the curves.

    ylim : tuple of shape (2,), default=None
        Defines minimum and maximum y-values plotted, e.g. (ymin, ymax).

    cv : int, cross-validation generator or an iterable, default=None
        Determines the cross-validation splitting strategy.
        Possible inputs for cv are:

          - None, to use the default 5-fold cross-validation,
          - integer, to specify the number of folds.
          - :term:`CV splitter`,
          - An iterable yielding (train, test) splits as arrays of indices.

        For integer/None inputs, if ``y`` is binary or multiclass,
        :class:`StratifiedKFold` used. If the estimator is not a classifier
        or if ``y`` is neither binary nor multiclass, :class:`KFold` is used.

        Refer :ref:`User Guide <cross_validation>` for the various
        cross-validators that can be used here.

    n_jobs : int or None, default=None
        Number of jobs to run in parallel.
        ``None`` means 1 unless in a :obj:`joblib.parallel_backend` context.
        ``-1`` means using all processors. See :term:`Glossary <n_jobs>`
        for more details.

    train_sizes : array-like of shape (n_ticks,)
        Relative or absolute numbers of training examples that will be used to
        generate the learning curve. If the ``dtype`` is float, it is regarded
        as a fraction of the maximum size of the training set (that is
        determined by the selected validation method), i.e. it has to be within
        (0, 1]. Otherwise it is interpreted as absolute sizes of the training
        sets. Note that for classification the number of samples usually have
        to be big enough to contain at least one sample from each class.
        (default: np.linspace(0.1, 1.0, 5))
    """
    if axes is None:
        _, axes = plt.subplots(1, 3, figsize=(20, 5))

    axes[0].set_title(title)
    if ylim is not None:
        axes[0].set_ylim(*ylim)
    axes[0].set_xlabel("Training examples")
    axes[0].set_ylabel("Score")

    train_sizes, train_scores, test_scores, fit_times, _ = learning_curve(
        estimator,
        X,
        y,
        cv=cv,
        n_jobs=n_jobs,
        train_sizes=train_sizes,
        return_times=True,
    )
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    fit_times_mean = np.mean(fit_times, axis=1)
    fit_times_std = np.std(fit_times, axis=1)

    # Plot learning curve
    axes[0].grid()
    axes[0].fill_between(
        train_sizes,
        train_scores_mean - train_scores_std,
        train_scores_mean + train_scores_std,
        alpha=0.1,
        color="r",
    )
    axes[0].fill_between(
        train_sizes,
        test_scores_mean - test_scores_std,
        test_scores_mean + test_scores_std,
        alpha=0.1,
        color="g",
    )
    axes[0].plot(
        train_sizes, train_scores_mean, "o-", color="r", label="Training score"
    )
    axes[0].plot(
        train_sizes, test_scores_mean, "o-", color="g", label="Cross-validation score"
    )
    axes[0].legend(loc="best")

    # Plot n_samples vs fit_times
    axes[1].grid()
    axes[1].plot(train_sizes, fit_times_mean, "o-")
    axes[1].fill_between(
        train_sizes,
        fit_times_mean - fit_times_std,
        fit_times_mean + fit_times_std,
        alpha=0.1,
    )
    axes[1].set_xlabel("Training examples")
    axes[1].set_ylabel("fit_times")
    axes[1].set_title("Scalability of the model")

    # Plot fit_time vs score
    fit_time_argsort = fit_times_mean.argsort()
    fit_time_sorted = fit_times_mean[fit_time_argsort]
    test_scores_mean_sorted = test_scores_mean[fit_time_argsort]
    test_scores_std_sorted = test_scores_std[fit_time_argsort]
    axes[2].grid()
    axes[2].plot(fit_time_sorted, test_scores_mean_sorted, "o-")
    axes[2].fill_between(
        fit_time_sorted,
        test_scores_mean_sorted - test_scores_std_sorted,
        test_scores_mean_sorted + test_scores_std_sorted,
        alpha=0.1,
    )
    axes[2].set_xlabel("fit_times")
    axes[2].set_ylabel("Score")
    axes[2].set_title("Performance of the model")

    return plt
```

```python
title = "Learning Curves (LinearRegression)"
estimator = LinearRegression()
#estimator = DecisionTreeClassifier(max_depth=3) 

plot_learning_curve(
     estimator, title, X_train, Y_train,n_jobs=4
)
```

Como ya habíamos visto, el desempeño de *LinearRegression* está en un rango medio. 

### Prueba con diferentes muestras.

```python
def Sampling(m, μ, σ):
    
    dist1 = norm(loc = μ[0], scale = σ[0])
    dist2 = norm(loc = μ[1], scale = σ[1])
    
    test_sample1 = dist1.rvs(size = m)
    test_sample2 = dist2.rvs(size = m)
    
    X = np.zeros(2*m)
    Y = np.zeros(2*m)
    X[:m] = test_sample1
    X[m:] = test_sample2
    Y[m:] = 1
    dic = {'X1':X, 'Y':Y} 
    data = pd.DataFrame(dic)

    X_train = data['X1'].to_numpy().reshape(-1,1)
    Y_train = data['Y']
    
    return X_train, Y_train
```

* $\mu_1=7.5$, $\mu_2=10$, $\sigma_1 = \sigma_2 = \sigma = 1 $

Para estas condiciones, usamos un entrenamiento con $30$ y con $100$ datos.

```python
s1_μ = [7.5, 10.0]
s1_σ = [1.0, 1.0]
X30_train, Y30_train = Sampling(30, s1_μ, s1_σ)
X100_train, Y100_train = Sampling(100, s1_μ, s1_σ)
```

```python
plot_learning_curve(
     estimator, title, X30_train, Y30_train, n_jobs=4
)
```

```python
plot_learning_curve(
     estimator, title, X100_train, Y100_train,n_jobs=4
)
```

Comparando las curvas de aprendizaje con el *training_set* se hace claro que el **aumento de la cantidad de datos** para entrenar mejoran la capacidad para aprender del modelo, pues para $30$ datos el *score* es cercano a $0.4$, mientras que para $100$ es cercano a $0.6$. 

* $\mu_1=7.5$, $\mu_2=7.0$, $\sigma_1 = \sigma_2 = \sigma = 2 $

```python
s2_μ = [7.5, 7.0]
s2_σ = [2.0, 2.0]
Xa200_train, Ya200_train = Sampling(200, s2_μ, s2_σ)

plot_learning_curve(
     estimator, title, Xa200_train, Ya200_train,n_jobs=4
)
```

* $\mu_1=7.5$, $\mu_2=7.0$, $\sigma_1 = \sigma_2 = \sigma = 0.5 $

```python
s3_μ = [7.5, 7.0]
s3_σ = [0.5, 0.5]
Xb200_train, Yb200_train = Sampling(200, s3_μ, s3_σ)

plot_learning_curve(
     estimator, title, Xb200_train, Yb200_train,n_jobs=4
)
```

En esta situación el rendimiento **en general es peor**, esto puede ser debido a que el clasificador usado no puede determinar de manera efectiva la frontera de decisión cuando las distribuciones están muy **sobrepuestas** y por tanto comparten gran parte de los datos suministrados. El hecho de que para distribuciones con medias fijas y **desviaciones estándar más grandes** (haciendo que se solapen más las distribuciones) el *score* tienda a **cero**, mientras que para las mismas medias y **desviaciones más pequeñas** el resultado sea cercano a $0.2$; es lo que lleva a pensar en esta conclusión.
