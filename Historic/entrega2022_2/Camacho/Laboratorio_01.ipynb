<h3>12.</h3> Determinar los datos [outlier](https://en.wikipedia.org/wiki/Outlier)  para la columna RadiusMean y eliminarlos del data frame, para ello construya un gráfico tipo [boxplot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.boxplot.html)

```
df.boxplot(column="RadiusMean", by='Diagnosis', sym = 'k.', figsize=(18,6))
```
¿Qué informacion podemos obtener de este tipo de graficos? 

En un problema de machine learning se debe elegir todas las columnas y construir un algoritmo que permite obtener la mejor calidad de los datos sobre todas las columnas.


````python
#boxplot de RadiusMean agrupandolo por B y M
data.boxplot(column="RadiusMean", by='Diagnosis', figsize=(18,6))
````

````python
#boxplot de RADIUSMEAN
data.boxplot(column="RadiusMean", sym = '.k', figsize=(18,6))
````

````python
# Las graficas boxplot nos da la información graficamente de los outliers, del rango intercuartil,
# de los interquartiles y de la mediana 
````

````python
#Quartiles
Q1 = np.percentile(data['RadiusMean'],25)
Q2 = np.percentile(data['RadiusMean'],50)
Q3 = np.percentile(data['RadiusMean'],75)


#RANGO intercuartil
RIC = Q3 - Q1
````

````python
data['RadiusMean'].describe()
````

````python
#frontera para determinar los outliers

maxi = Q3 + 1.5*RIC
mini = Q1 - 1.5*RIC
````

````python
#Outliers de la columna RadiusMean, los que estan por encima de max o debajo de min
data_wout = data.drop(data[(data['RadiusMean']>maxi) | (data['RadiusMean']<mini)].index)
data_wout
````

````python
#Filtrando solo los datos que no son Outliers en la columna RadiusMean
data_wout = data[(data['RadiusMean'] <= maxi) & (data['RadiusMean'] >= mini)]
````

````python
data_wout
````

# Laboratorio 1.1: Series de tiempo

<h3>1. Leer el archivo de datos</h3>

````python
url = "http://berkeleyearth.lbl.gov/auto/Global/Land_and_Ocean_complete.txt"
````

````python
df = np.loadtxt(url,comments='%')
````

<h3>2. Construir un dataframe con las columnas:</h3>


```
cols=['RadiusMean', 'TextureMean', 'PerimeterMean', 'AreaMean','SmoothnessMean', 'CompactnessMean', 'ConcavityMean','ConcavePointsMean']
```

````python
data_t = pd.DataFrame(df,columns = ["Year","Month","MonthlyA","MonthlyUnc","AnnuealA","AnnuealUnc",
                                  "FiveA","AnnuealUnc","TenA","AnnuealUnc","TwentyA","AnnuealUnc"])
data_t
````

<h3>3. Construir una nueva columna tipo string llamada Date con la informacion del Year y Month</h3>

````python
string1 = list(map(lambda s: str(int(s)), data_t['Year']))
string2 = list(map(lambda s: str(int(s)), data_t['Month'])) 
````

````python
data_t['Date'] = [ string1[i]+string2[i] for i in range(len(string1))] 
data_t
````

<h3>4. Emplear el commando [datetime](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) para dar el formato tipo date  a la nueva columna Date.</h3>

````python
data_t['Date'] = pd.to_datetime(data_t['Date'],format='%Y%m')
````

````python
data_t
````

<h3>5. Emplear el metodo groupby and grouper para tomar los datos con un periodicidad mensual. Notese que freq permite variar la periodicidad con la que se quieren mostrar los datos. Pruebe para freq="2M", "3M", "Y", "H".</h3>

````python
#No hay registros de horas
data_t.groupby(pd.Grouper(key = 'Date', freq='H')).mean()
````

````python
T_serie_M = data_t.groupby(pd.Grouper(key = 'Date', freq='M')).mean()
T_serie_2M = data_t.groupby(pd.Grouper(key = 'Date', freq='2M')).mean()
T_serie_3M = data_t.groupby(pd.Grouper(key = 'Date', freq='3M')).mean()
T_serie_6M = data_t.groupby(pd.Grouper(key = 'Date', freq='6M')).mean()
T_serie_Y = data_t.groupby(pd.Grouper(key = 'Date', freq='Y')).mean()

T_series = [T_serie_M, T_serie_2M, T_serie_3M, T_serie_6M, T_serie_Y]
tt = ['Mes','2 Meses','3 Meses','6 Meses','Año']
````

<h3>6. Realice un gráfico del tiempo como función de  MonthlyA.</h3>

````python
i=0
for data in T_series:
    i+=1
    plt.figure(figsize=(16,4))
    plt.plot(data['MonthlyA'], label =f'{tt[i-1]}')
    plt.legend()
    plt.show()
````

<h3>7. ¿Que puede concluir?</h3>

A través de las anteriores gráficas se puede observar que la variable `MonthlyA` tiene un comportamiento poco predecible a lo largo de los meses, ya que no sigue un comportamiento constante, esto para la epoca entre los años (1850 a $\sim$ 1920) aproximadamente, donde presenta varios picos de minimos y máximos, y luego del año $\sim$ 1920, a pesar de que el comportamiento por mes sigue fluctuando, la variable va aumentando ya que por ejemplo para los último años (2019-2021) se observa que sus valores son los máximos en todo el DataFrame.
