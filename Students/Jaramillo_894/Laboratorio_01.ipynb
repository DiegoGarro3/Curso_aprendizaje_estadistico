```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

def Pascal(df):
    # Convierte al formato PascalCase
    cols1 = [i.replace("_"," ").split() for i in df.columns]
        # Reemplazar "_" por " "

    cols2 = [[i.capitalize() for i in j] for j in cols1]
        # Capitalize a cada palabra de cada lista de cols1

    cols3 = ["".join(i) for i in cols2]
        # Juntar todas las palabras sin espacio

    return cols3 # Retorna la lista con los nuevos nombres

def numero(x):
    # Determina si x se puede convertir a un numero
    try:
        float(x)
        return True
    except:
        return False
```

## Determinar los datos outlier para la columna RadiusMean y eliminarlos del data frame, para ello construya un gráfico tipo boxplot

```python
df.boxplot(column="RadiusMean", by='Diagnosis', sym = 'k.', figsize=(18,6))
```

```python
df_norm.boxplot(column="RadiusMean", by='Diagnosis', sym = 'k.', figsize=(18,6))
plt.title("Grafico datos normalizados para RadiusMean")
```

```python
#  https://medium.com/@ameya_shukla/percentiles-box-plots-their-intersection-with-data-science-25e8204e2683

for i in df.columns:
    # Filtro para los datos outlier
    if numero(df_norm[i][0]):

        std_B = df_norm[i][(df_norm["Diagnosis"]=="B")].std()
        mean_B = df_norm[i][(df_norm["Diagnosis"]=="B")].mean()

        Q1_B = df_norm[i][(df_norm["Diagnosis"]=="B")].quantile(0.25)
        Q3_B = df_norm[i][(df_norm["Diagnosis"]=="B")].quantile(0.75)
        IQR_B = Q3_B - Q1_B


        std_M = df_norm[i][(df_norm["Diagnosis"]=="M")].std()
        mean_M = df_norm[i][(df_norm["Diagnosis"]=="M")].mean()

        Q1_M = df_norm[i][(df_norm["Diagnosis"]=="M")].quantile(0.25)
        Q3_M = df_norm[i][(df_norm["Diagnosis"]=="M")].quantile(0.75)
        IQR_M = Q3_M - Q1_M

        df_norm[i][(df_norm["Diagnosis"]=="B") & (abs( mean_B - df_norm[i])>1.5*IQR_B)] = mean_B
        df_norm[i][(df_norm["Diagnosis"]=="M") & (abs( mean_M - df_norm[i])>1.5*IQR_M)] = mean_M

```

```python
df_norm.boxplot(column="RadiusMean", by='Diagnosis', sym = 'k.', figsize=(18,6))
plt.title("Grafico datos normalizados para RadiusMean con datos filtrados")
```

### Para los datos asociados a los meses de 2023, construya un histograma para cada mes.

```python
asociados_mes = dt_23.groupby(pd.Grouper( freq='1M'))
```

```python
#######
fig, ax = plt.subplots()
for month, group_data in asociados_mes:
    ax.hist(group_data['DiffPrice'], bins=20, alpha=0.5, label=month.strftime('%B %Y'))

ax.set_xlabel('Valor')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma por Mes')
ax.legend()

```

```python
dt_23['MesNum'] = dt_23.index.month        # Crea una columna con el numero del mes
dt_23['Mes'] = dt_23.index.strftime('%B')  # Crea una columna con el mes

dt_23
```

```python
#white, dark, whitegrid, darkgrid, ticks
#sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# Initialize the FacetGrid object
pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
g = sns.FacetGrid(dt_23, row="Mes", hue="Mes", aspect=5, height=1, palette=pal)


# Draw the densities in a few steps
g.map(sns.histplot, "DiffPrice", clip_on=False,
      fill=True, alpha=1, linewidth=1.5)
g.map(sns.histplot, "DiffPrice", clip_on=False, color="w", lw=2)

# passing color=None to refline() uses the hue mapping
g.refline(y=0, linewidth=2, linestyle="-", color=None, clip_on=False)


# Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)


g.map(label, "DiffPrice")


# Set the subplots to overlap
g.figure.subplots_adjust(hspace=0.1)
g.set(xlim=(-0.005, 0.005))
#plt.xlim(-0.005, 0.005)

# Remove axes details that don't play well with overlap
g.set_titles("")
g.set(yticks=[], ylabel="")
g.despine(bottom=True, left=True)
```
