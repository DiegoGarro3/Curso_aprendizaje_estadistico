```python
def normalizeSeries(x):
  mean=x.mean()
  std=x.std()
  return (x-mean)/std

def normalizeDF(df):
  df=df.select_dtypes(include=[np.number])

  for column in df.columns:
    df[column]=normalizeSeries(df[column])
  return df
```

```python
dfNormalized=normalizeDF(df)
```

## Eliminaci√≥n de outliers

```python
def outlier_delete(df, name_col):

  Q1=df[name_col].quantile(0.25)
  Q3=df[name_col].quantile(0.75)
  IQR=Q3-Q1
  IQR
  condition= ~( (df[name_col]> (Q3+1.5*IQR) ) | (df[name_col] <(Q1-1.5*IQR)  ) )
  return df[condition].reset_index(drop=True)
```

```python
dfClean=outlier_delete(df, 'RadiusMean')
```

## Dataset Series de tiempo

## 1)

```python
df=pd.read_csv('http://berkeleyearth.lbl.gov/auto/Global/Land_and_Ocean_complete.txt',
               comment='%', sep='\s+',header=None)
```

## 2)

```python
df.head()
```

```python
df.columns=['Year','Month','Monthly Anomaly', 'Unc','Annual Anomaly', 'Unc',
            'Five-year Anomaly', 'Unc', 'Ten-year Anomaly', 'Unc','Twenty-year Anomaly' ,'Unc']
```

```python
df.head()
```

```python
def Month_function(x):
  if len(str(x) )==1:
    x='0'+str(x)
  return x

df['Month']=df['Month'].apply(Month_function)
```

```python
df.head(2)
```

```python
df['Date']=df['Year'].apply(str)+'-'+df['Month'].apply(str)
```

```python
df.head()
```

```python
df["Date"]=pd.to_datetime(df['Date'], infer_datetime_format=True)
```

## 3) y 4)

```python
df2M=df.groupby(pd.Grouper(key='Date', freq='2M') )['Monthly Anomaly'].mean()
df3M=df.groupby(pd.Grouper(key='Date', freq='3M') )['Monthly Anomaly'].mean()
df1Y=df.groupby(pd.Grouper(key='Date', freq='1Y') )['Monthly Anomaly'].mean()
df1H=df.groupby(pd.Grouper(key='Date', freq='H') )['Monthly Anomaly'].mean()

plt.figure(figsize=(10,4))
plt.title("Average of two month of monthly anomaly temperature ")
df2M.plot()
plt.show()
plt.close()

plt.figure(figsize=(10,4))
plt.title("Average of three month of monthly anomaly temperature ")
df3M.plot()
plt.show()
plt.close()

plt.figure(figsize=(10,4))
plt.title("Average of one year of monthly anomaly temperature")
df1Y.plot()
plt.show()
plt.close()

plt.figure(figsize=(10,4))
plt.title("Average of one hour of monthly anomaly temperature ")
df1H.plot()
plt.show()
plt.close()
```
