## Entrenamiento con sklearn (Logistic Regression)

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
```

```python
for train_index, test_index in split.split(df, df["Target"]):
    strat_train_set = df.loc[train_index] 
    strat_test_set = df.loc[test_index]
```

```python
# setosa=0, versicolor=1, virginica=2
iris.target_names
```

```python
# Separar los datos de setosa y no-setosa 
# Entonces se hace setosa=1 y no-setosa=0
df1_train = strat_train_set.copy()
df1_test = strat_test_set.copy()
```

```python
# Poniendo setosa=1 Y no-setosa=0
df1_train['Target'] = df1_train['Target'].apply(lambda x: 1 if x==0 else 0)
df1_test['Target'] = df1_test['Target'].apply(lambda x: 1 if x==0 else 0)
```

```python
df1_train
```

```python
# Separando los features y labels para el entrenamiento y el testeo
X1_train = df1_train[['SepalLength_cm', 'SepalWidth_cm', 'PetalLength_cm', 'PetalWidth_cm']].to_numpy()
y1_train = df1_train['Target'].to_numpy()

X1_test = df1_test[['SepalLength_cm', 'SepalWidth_cm', 'PetalLength_cm', 'PetalWidth_cm']].to_numpy()
y1_test = df1_test['Target'].to_numpy()
```

```python
from sklearn.linear_model import LogisticRegression

clf1 = LogisticRegression()
```

```python
X1_train.shape
```

```python
y1_train.shape
```

```python
clf1.fit(X1_train, y1_train)
```

```python
# Score del training
clf1.score(X1_train, y1_train)
```

```python
# Score del test
clf1.score(X1_test, y1_test)
```

```python
sns.pairplot(data=df, hue='Target', palette='dark')
```

A través de la gráfica de arriba se observa que la separación entre las tres clases es visible, pero que la clase `setosa` = `0` (la azul) es la que presenta una mayor separación y por tanto una mayor facilidad para diferenciarla de las otras dos clases. Por tanto la anterior frontera con la regresión logistica arroja un `score=1.0`.

Dado X = [4.9,5.0, 1.8, 0.3] asociados a todas las caracteristicas, ¿cuál es la probabilidad de que la flor sea setosa, versicolor o virginica? 


```python
# setosa=0, versicolor=1, virginica=2
X_prueba = [[4.9,5.0, 1.8, 0.3]] 
```

```python
clf.predict_proba(X_prueba)
```

Entonces, una planta con las caracteristicas de `X_prueba = [[4.9,5.0, 1.8, 0.3]] ` tiene una probabilidad del $\sim 0.992 \sim 99\%$ de ser de clase `setosa = 0`
