```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

##1.
Leer el archivo de datos

```python
url = "http://berkeleyearth.lbl.gov/auto/Global/Land_and_Ocean_complete.txt"
```

```python
data = np.loadtxt(url,comments ="%")
```

##2.
Construir un dataframe con las columnas:

```python
df = pd.DataFrame(data,columns = ["Year","Month","MonthlyA","MonthlyUnc","AnnuealA","AnnuealUnc","FiveA","AnnuealUnc","TenA","AnnuealUnc","TwentyA","AnnuealUnc"])
#columns = ["Year", "Month",  "Monthly_Anomaly", "Monthly_Unc.",   "Annual_Anomaly", "Annual_Unc.",   "Five-year_Anomaly", "Five-year_Unc.",   "Ten-year_Anomaly", "Ten-year_Unc.",   "Twenty-year_Anomaly", "Twenty-year_Unc."])
```

##3. 
Construir una nueva columna tipo string llamada Date con la informacion del Year y Month

```python
c = []
d= "_"
for i in range(len(df["Year"])):
  #for j in range(len(df["Month"])):
  a = str(int(df["Year"].iloc[i]))
  b = str(int(df["Month"].iloc[i]))
  c.append(a+b)
```

##6.

Realice un gr√°fico del tiempo como funci√≥n de MonthlyA.

```python
plt.figure(dpi=100)
plt.plot(df["MonthlyA"],df["Date"],".",ms=0.8)
plt.xlabel(r"$MonthlyA$")
plt.ylabel(r"$Date$")
plt.title("Gr√°fico del tiempo como funci√≥n de MonthlyA")
plt.legend()
plt.grid()
```

##7.

¬øQue puede concluir?

Si sacamos una gr√°fica del promedio de la temperatura a lo largo del a√±o por cada a√±o, obtenemos

```python
plt.figure(dpi=100)
plt.plot(np.array(df.groupby(pd.Grouper(key='Date', freq='Y')).MonthlyA.mean()),np.array(df.groupby(pd.Grouper(key='Date', freq='Y')).Year.mean()),"o",ms=0.8)
plt.xlabel(r"$MonthlyA$")
plt.ylabel(r"$Date$")
plt.title("Gr√°fico del tiempo como funci√≥n de MonthlyA")
plt.legend()
plt.grid()
```

Como podemos observar en ambas gr√°ficas a lo largo del tiempo la temperatura global de la tierra ha ido aumentando, notandose un aumento mucho m√°s acelerado a partir de los a√±os 80's pareciendose mucho la distribuci√≥n a una logar√≠timica. Esto muestra claramente que el calentamiento global ha sido directamente proporcional al aumento de tecnolog√≠as, producci√≥n global y tasa de natalidad en la tierra, donde aproximadamente en estos a√±os es donde comienza a acelerar mucho m√°s hasta el d√≠a de hoy.
