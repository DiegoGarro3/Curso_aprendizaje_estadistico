```python
%pylab inline
import pandas as pd
from mpl_toolkits import mplot3d
from random import sample
```

Una forma de generalizar la regresión lineal  para $N$ características es

$$J = \frac{1}{2m}\sum_{i}^{m}\left[ \theta_0 + \theta_1 x_i^{1} + \theta_2 x_i^{2} + \cdots + \theta_N x_i^{N} -y_i\right]^2$$

donde 

$$\frac{\partial J}{\partial \theta_s} = \frac{1}{m}\sum_{i}^{m}\left[ \theta_0 + \theta_1 x_i^{1} + \theta_2 x_i^{2} + \cdots + \theta_N x_i^{N} -y_i\right]x_{i}^{s}$$

y por tanto en el algoritmo de GD:

$$\Theta_{n+1} = \Theta_{n} - \alpha\nabla J$$

### Ejemplo.

A modo de ilustración, para el plano $y = 2x^1+3x^2$ se generarán datos y buscaremos obtener los coeficientes de la ecuación del plano a partir de estos datos.

```python
def Plano(x,y):
    return 2*x+3*y

#Rangos.
xMin, xMax, yMin, yMax = [0., 1., 0., 1.]

n = 1000

X_data = [[(xMax-xMin)*np.random.random()+xMin, (yMax-yMin)*np.random.random()+yMin] for i in range(n)]
y_data = [ Plano(pair[0], pair[1]) for pair in X_data]
```

```python
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111 ,projection='3d')

ax.scatter(np.transpose(X_data)[0], np.transpose(X_data)[1], y_data)
plt.show()
```

Dentro de estos puntos vamos a seleccionar una **muestra** que servirá como *training_set*.

```python
m = 20
X_sample = sample(X_data, m)
y_sample = Plano(np.transpose(X_sample)[0], np.transpose(X_sample)[1])
```

```python
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111 ,projection='3d')

ax.scatter(np.transpose(X_data)[0], np.transpose(X_data)[1], y_data, c='red', alpha=0.3)
ax.scatter(np.transpose(X_sample)[0], np.transpose(X_sample)[1], y_sample, c='green', alpha=1.)
plt.show()
```

Para hallar los valores de $\theta_0$, $\theta_1$ y $\theta_2$ usamos la función de coste de manera vectorial

$$J = \frac{1}{2m}\sum_{i}^{m}(\Theta^T \tilde{X}-y)^2$$

con 

$$\Theta^T = \pmatrix{\theta_0 & \theta_1 & \theta_2} \qquad \text{ y }\qquad \tilde{X} = \pmatrix{1 & 1 & \dots & 1 \\
x^{(1)}_1 & x^{(2)}_1 & \dots & x^{(m)}_1 \\ x^{(1)}_2 & x^{(2)}_2 & \dots & x^{(m)}_2}$$ 

Apatir de aquí se encuentra que el gradiente de $J$ es:

$$\nabla J = (\Theta^T\tilde{X}-y)\tilde{X}$$

Estas funciones son implementadas en el siguiente código.

```python
def J(Θ, X_sample, y_sample):
    nPar = len(Θ)
    nFeat = len(y_sample)
    df = pd.DataFrame({'Y':y_sample, 'X1':np.transpose(X_sample)[0], 'X2':np.transpose(X_sample)[1]})
    df['ones'] = np.ones(nFeat)
    #Construcción de la matriz X.
    X = df[['ones', 'X1', 'X2']].values
    return np.dot(np.dot(Θ, X.transpose())-y_sample, np.dot(Θ, X.transpose())-y_sample)

def DJ(Θ, X_sample, y_sample):
    nPar = len(Θ)
    nFeat = len(y_sample)
    df = pd.DataFrame({'Y':y_sample, 'X1':np.transpose(X_sample)[0], 'X2':np.transpose(X_sample)[1]})
    df['ones'] = np.ones(nFeat)
    X = df[['ones', 'X1', 'X2']].values
    return np.dot(np.dot(Θ, X.transpose())-y_sample ,X)
```

A continuación se inicializan los parámetros $\Theta$ de manera aleatoria y se define la tasa de aprendizaje.

```python
ilim, flim = [-3, 3] 
Θp = np.random.rand(3)*(flim-ilim)+ilim
α = 0.01
```

```python
#Parámetros iniciales.
Θp
```

El gradiente descendente queda implementado de manera sencilla a través de un ciclo con `k` iteraciones como:

$$\Theta_{k} = \Theta_{k-1}-\alpha\nabla J$$

```python
k = 1500
for i in range(k):
    Θp -= α*DJ(Θp, X_sample, y_sample)
```

```python
print("Los parámetros de ajuste son {:.2}, {:.5}, {:.6}".format(*Θp))
```

En efecto, la regresión por GD ha devuelto los **valores correctos** para los coeficientes en la ecuación del **plano** que se definió.
