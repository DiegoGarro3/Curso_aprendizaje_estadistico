# 1. Contrucción y gráfica de dos distribuciones

````python
x = np.linspace(-5, 15,1000)

#Parámetros de las distribuciones
mean1 = 5
mean2=7.5
desv1 = 2.5
desv2=1.5
f1 = stats.norm(loc = mean1, scale = desv1)
f2=stats.norm(loc=mean2, scale=desv2)
````

````python

plt.figure(figsize=(12,8))
plt.title("Distribuciones", fontsize="25")
plt.plot(x,f1.pdf(x), label="$\\mu=5$, $\\sigma=2.5$") #Se grafican
plt.plot(x,f2.pdf(x), label="$\\mu=7.5$, $\\sigma=1.5$")
plt.legend(loc="upper right", fontsize="15")
````

# 2. Frontera analítica 

Cada una de las distribuciones cumple ---respectivamente--- que:

$f_{1}=\frac{1}{\sigma_{1}\sqrt{2\pi}}e^{-\frac{1}{2}\bigg(\frac{x-\mu_{1}}{\sigma_{1}}\bigg)^2}$ y $f_{2}=\frac{1}{\sigma_{2}\sqrt{2\pi}}e^{-\frac{1}{2}\bigg(\frac{x-\mu_{2}}{\sigma_{2}}\bigg)^2}$

La intersección entre ellas y ergo su frontera es allí donde ambas resultan iguales:

$\frac{1}{\sigma_{1}}e^{-\frac{1}{2}\bigg(\frac{x-\mu_{1}}{\sigma_{1}}\bigg)^{2}}=\frac{1}{\sigma_{2}}e^{-\frac{1}{2}\bigg(\frac{x-\mu_{2}}{\sigma_{2}}\bigg)^{2}}$

Entonces,

$\ln\bigg(\frac{\sigma_{2}}{\sigma_{1}}\bigg)=\frac{1}{2}\bigg[\big(\frac{x-\mu_{1}}{\sigma_{1}}\big)^{2}-\big(\frac{x-\mu_{2}}{\sigma_{2}}\big)^{2}\bigg]$

de donde finalmente resulta la ecuación cuadrática:

$0=\big(\sigma_{2}^{2}-\sigma_{1}^2\big)x^2-2\big(\mu_{1}\sigma_{2}^2-\mu_{2}\sigma_{1}^2\big)x+\mu_{1}^{2}\sigma_{2}^2-\mu_{2}^2\sigma_{1}^2-\sigma_{1}^2\sigma_{2}^2\ln\bigg(\frac{\sigma_{2}^2}{\sigma_{1}^2}\bigg)$.


De esta ecuación ---para los valores particulares asignados--- surgen dos soluciones, $x\sim3.26$ y $x\sim-32.15$, de las que evidentemente se toma la primera de ellas.



# 3 y 4. Muestras aleatorias

````python
m=100 #Número de valores aleatorios a generar. Datos de entrenamiento
n=25 #Datos de prueba
random_1 = f1.rvs(m) #Primera distribución
random_2 = f2.rvs(m) #Segunda distribución 
test_1=f1.rvs(n) #Datos de prueba primera distribución
test_2=f2.rvs(n) #Datos de prueba segunda distribución

#Se grafican los datos de entrenamiento 
plt.figure(figsize=(12,8))
plt.title("Distribuciones de entrenamiento", fontsize="25")
plt.hist(random_1, density=True)
plt.plot(x, f1.pdf(x))
plt.hist(random_2, density=True)
plt.plot(x, f2.pdf(x))
plt.show()
````

# 5 y 6. Creación de los DataFrame de entrenamiento y prueba

````python
X=np.zeros(2*m)#Arreglo de ceros, 200 entradas
Y=np.zeros(2*m)

X_test=np.zeros(2*n) #Arreglo de ceros para datos en entrenamiento. 50 entradas
Y_test=np.zeros(2*n)

X[0:m]=random_1 #Distribución 1 para las priemras cien entradas
X[m:2*m]=random_2 #Distribución 2 para las siguientes 100 entradas
X_test[0:n]=test_1
X_test[n:2*n]=test_2

Y[0:m]=np.zeros(m) #Etiqueta 0 para las primeras 100 entradas
Y[m:2*m]=np.ones(m) #Etiqueta 1 para las siguientes 100 entradas
Y_test[0:n]=np.zeros(n)
Y_test[n:2*n]=np.ones(n)
````

````python
data_muestra=np.array([Y,X]).T
data_test=np.array([Y_test,X_test]).T
columns_muestra=['Y', 'X']
columns_test=['Y_test', 'X_test']

df_muestra=pd.DataFrame(data=data_muestra, columns=columns_muestra) #DataFrame entrenamiento
df_test=pd.DataFrame(data=data_test, columns=columns_test) #DataFrame prueba
df_muestra
````

# Cruvas de aprendizaje

````python
def plot_learning_curve(
    estimator,
    title,
    X,
    y,
    axes=None,
    ylim=None,
    cv=None,
    n_jobs=None,
    train_sizes=np.linspace(0.1, 1.0, 5),
):
    """
    Generate 3 plots: the test and training learning curve, the training
    samples vs fit times curve, the fit times vs score curve.

    Parameters
    ----------
    estimator : estimator instance
        An estimator instance implementing `fit` and `predict` methods which
        will be cloned for each validation.

    title : str
        Title for the chart.

    X : array-like of shape (n_samples, n_features)
        Training vector, where ``n_samples`` is the number of samples and
        ``n_features`` is the number of features.

    y : array-like of shape (n_samples) or (n_samples, n_features)
        Target relative to ``X`` for classification or regression;
        None for unsupervised learning.

    axes : array-like of shape (3,), default=None
        Axes to use for plotting the curves.

    ylim : tuple of shape (2,), default=None
        Defines minimum and maximum y-values plotted, e.g. (ymin, ymax).

    cv : int, cross-validation generator or an iterable, default=None
        Determines the cross-validation splitting strategy.
        Possible inputs for cv are:

          - None, to use the default 5-fold cross-validation,
          - integer, to specify the number of folds.
          - :term:`CV splitter`,
          - An iterable yielding (train, test) splits as arrays of indices.

        For integer/None inputs, if ``y`` is binary or multiclass,
        :class:`StratifiedKFold` used. If the estimator is not a classifier
        or if ``y`` is neither binary nor multiclass, :class:`KFold` is used.

        Refer :ref:`User Guide <cross_validation>` for the various
        cross-validators that can be used here.

    n_jobs : int or None, default=None
        Number of jobs to run in parallel.
        ``None`` means 1 unless in a :obj:`joblib.parallel_backend` context.
        ``-1`` means using all processors. See :term:`Glossary <n_jobs>`
        for more details.

    train_sizes : array-like of shape (n_ticks,)
        Relative or absolute numbers of training examples that will be used to
        generate the learning curve. If the ``dtype`` is float, it is regarded
        as a fraction of the maximum size of the training set (that is
        determined by the selected validation method), i.e. it has to be within
        (0, 1]. Otherwise it is interpreted as absolute sizes of the training
        sets. Note that for classification the number of samples usually have
        to be big enough to contain at least one sample from each class.
        (default: np.linspace(0.1, 1.0, 5))
    """
    if axes is None:
        _, axes = plt.subplots(1, 3, figsize=(20, 5))

    axes[0].set_title(title)
    if ylim is not None:
        axes[0].set_ylim(*ylim)
    axes[0].set_xlabel("Training examples")
    axes[0].set_ylabel("Score")

    train_sizes, train_scores, test_scores, fit_times, _ = learning_curve(
        estimator,
        X,
        y,
        cv=cv,
        n_jobs=n_jobs,
        train_sizes=train_sizes,
        return_times=True,
    )
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    fit_times_mean = np.mean(fit_times, axis=1)
    fit_times_std = np.std(fit_times, axis=1)

    # Plot learning curve
    axes[0].grid()
    axes[0].fill_between(
        train_sizes,
        train_scores_mean - train_scores_std,
        train_scores_mean + train_scores_std,
        alpha=0.1,
        color="r",
    )
    axes[0].fill_between(
        train_sizes,
        test_scores_mean - test_scores_std,
        test_scores_mean + test_scores_std,
        alpha=0.1,
        color="g",
    )
    axes[0].plot(
        train_sizes, train_scores_mean, "o-", color="r", label="Training score"
    )
    axes[0].plot(
        train_sizes, test_scores_mean, "o-", color="g", label="Cross-validation score"
    )
    axes[0].legend(loc="best")

    # Plot n_samples vs fit_times
    axes[1].grid()
    axes[1].plot(train_sizes, fit_times_mean, "o-")
    axes[1].fill_between(
        train_sizes,
        fit_times_mean - fit_times_std,
        fit_times_mean + fit_times_std,
        alpha=0.1,
    )
    axes[1].set_xlabel("Training examples")
    axes[1].set_ylabel("fit_times")
    axes[1].set_title("Scalability of the model")

    # Plot fit_time vs score
    fit_time_argsort = fit_times_mean.argsort()
    fit_time_sorted = fit_times_mean[fit_time_argsort]
    test_scores_mean_sorted = test_scores_mean[fit_time_argsort]
    test_scores_std_sorted = test_scores_std[fit_time_argsort]
    axes[2].grid()
    axes[2].plot(fit_time_sorted, test_scores_mean_sorted, "o-")
    axes[2].fill_between(
        fit_time_sorted,
        test_scores_mean_sorted - test_scores_std_sorted,
        test_scores_mean_sorted + test_scores_std_sorted,
        alpha=0.1,
    )
    axes[2].set_xlabel("fit_times")
    axes[2].set_ylabel("Score")
    axes[2].set_title("Performance of the model")

    return plt
````

````python
title = "Learning Curves (LinearRegression)"
plot_learning_curve(LinearRegression(), title, X.reshape(-1,1), Y.reshape(-1,),  ylim=None,  n_jobs=None, train_sizes=np.linspace(0.1,1,50))
````

# 10. Curvas de aprendizaje para distintos valores

````python
def  Aleatorio (loc, scale, m): #Función para generar los valores aleatorios
    f=stats.norm(loc=loc, scale = scale) 
    return f.rvs(m)

N=[30,100,200,200] #Arreglo con los distintos  números de valores a usar
mu_1=list(np.ones(4)*7.5) #Media de cada una de las primeras distribuciones
mu_2=[10,10,7,7] #Media de cada una de las segundas distribuciones
sigma_1=[1,1,2,0.5] #DEsviación estándar priemras distribuciones
sigma_2=sigma_1

XX=[] #Cada entrada de este arreglo es un arreglo con las distribuciones 1 y 2 para cada N
YY=[] 
for i in range(len(N)):
    
    XX1=list(Aleatorio(mu_1[i],sigma_1[i],N[i])) #Valores de la priemra distribución
    XX2=list(Aleatorio(mu_2[i],sigma_2[i],N[i])) #Valores segunda distribución
    YY1=[0 for i in range(N[i])]
    YY2=[1 for i in range(N[i])]
    
    XX.append(XX1+XX2) #Concatena los datos de la primera y segunda distribución y los asocia a la entrada i-ésima de la lista
    YY.append(YY1+YY2)


classifier=LinearRegression() #Regresion lineal
title="Learning Curves"
for i in range(4):#Se grafican las curvas de aprendizaje
    plot_learning_curve(
        classifier, title+str(" ")+str(N[i]), np.array(XX[i]).reshape(-1,1), np.array(YY[i]).reshape(-1,), axes=None, ylim=None, cv=None,  n_jobs=None, train_sizes=np.linspace(0.1,1.0,50),)
````
