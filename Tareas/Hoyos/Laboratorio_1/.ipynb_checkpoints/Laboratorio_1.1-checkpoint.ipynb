# 1. Lectura de datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt("http://berkeleyearth.lbl.gov/auto/Global/Land_and_Ocean_complete.txt",  comments='%')
columns = ["Year","Month","MonthlyA","MonthlyUnc","AnnualA","AnnualUnc","FiveA","FiveUnc","TenA","TenUnc","TwentyA","TwentyUnc"]
```

# 2. Construcción del dataframe

```python
df=pd.DataFrame(data=data,columns=columns)
df #Se visualiza
```

# 3. Construcción Date

```python
df['YearStr']=list(map(lambda x: str(int(x)), df['Year'])) #Convierte los datos de la columna Year en strings y los guarda en YearStr
df['MonthStr']=list(map(lambda x: str(int(x)), df['Month'])) #Convierte los datos de la columna Month en strings y los guarda en MonthStr
df['Date']=df['YearStr']+df['MonthStr'] #Única columna con la información de año y mes
```

```python
df #Se visualizan los resultados
```

# 4. Formato date-time

```python
df["Date"] = pd.to_datetime(df["Date"],format='%Y%m')
```

```python
df #Se visualiza de nuevo
```

# 5. Datos agrupados por periodicidad


```python
T_serie0 = df.groupby(pd.Grouper(key='Date', freq='M')).mean() #Periodicidad mensual
T_serie0 #Se imprime el dataframe con la agrupación deseada
```

```python
T_serie = df.groupby(pd.Grouper(key='Date', freq='2M')).mean() #Periodiciadad bimensual
T_serie
```

```python
T_serie = df.groupby(pd.Grouper(key='Date', freq='3M')).mean() #Periodicidad trimestral
T_serie
```

# 6. Gráfico

```python

plt.figure(figsize=(18,5))
plt.title("Anomalías mensuales de la temperatura terrestre", fontsize="30")
plt.plot(T_serie0.index, T_serie0['MonthlyA'])
plt.ylabel("Anomalía mensual", fontsize="15")
plt.xlabel("Fecha", fontsize="15")
plt.grid()
```

Del gráfico precendente es claro que las anomalías de la temperatura terrestre respecto al promedio de enero 1951-diciembre 1980 han ido en aumento. De dataframe también se ve que las incertidumbres de estas anomalías tienden a decrecer. También puede verse que además del aumentó hay una tendencia que muestra que este seguirá sucediendo, la curva no parece tener un comportamiento asintótico, por lo menos no en el intervalo considerado.
